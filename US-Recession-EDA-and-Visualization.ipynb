# **DATA VISUALIZATION**

##  1. Check Recession Frequency

```python
# Countplot of recession occurrences, Seaborn's countplot(), we don't explicitly mention that it's a bar graph because countplot() automatically creates a bar chart
plt.figure(figsize=(6,4))
sns.countplot(x=df1['Recession'], palette="coolwarm")
plt.xlabel("Recession (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Recession Occurrences in Dataset")
plt.grid()
plt.show()
```

# 2. Analyze GDP Trends During Recessions

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df1['Recession'], y=df1['GDP'], palette="coolwarm")
plt.xlabel("Recession (0 = No, 1 = Yes)")
plt.ylabel("GDP Growth (%)")
plt.title("GDP Distribution During Recessions vs. Non-Recessions")
plt.show()
```

If GDP drops significantly during recessions, it confirms its importance as a leading indicator.

# Housing Index During Recession vs. Non-Recession

```python
plt.figure(figsize=(5,5))

# Calculate the average Housing Index for Recession (1) and Non-Recession (0)
housing_avg = df.groupby('Recession')['Housing_Index'].mean()

# Define labels and colors
labels = ["No Recession (0)", "Recession (1)"]
colors = ["skyblue", "salmon"]

# Create the pie chart
plt.pie(housing_avg, labels=labels, autopct="%1.1f%%",startangle=90, explode=(0, 0.1), colors=colors, shadow=True)

plt.title("Average Housing Index During Recession vs. Non-Recession")
plt.show()
```

Computes the average Housing Index for Recession = 0 and 1.

# Average Interest Rate During Recession vs. Non-Recession

```python
plt.figure(figsize=(8,5))

# Calculate the average interest rate for Recession (1) and Non-Recession (0)
interest_avg = df.groupby('Recession')['Rate'].mean()

# Define colors for each bar
colors = ["violet", "blue"]  # Blue for no recession, Red for recession

# Create a bar plot with different colors
sns.barplot(x=interest_avg.index, y=interest_avg.values, palette=colors)

# Labeling the graph
plt.xlabel("Recession (0 = No, 1 = Yes)")
plt.ylabel("Average Interest Rate (%)")
plt.title("Average Interest Rate During Recession vs. Non-Recession")

plt
```

# Compare Key Economic Indicators During Recessions

```python
recession_data = df1[df1['Recession'] == 1]  # Filters rows where the Recession column is 1
non_recession_data = df1[df1['Recession'] == 0] #Filters rows where the Recession column is 0

# Average values of indicators
print("During Recession:")
print(recession_data[['GDP', 'INDPRO', 'CPI', 'Housing_Index']].mean())

print("\nDuring Non-Recession:")
print(non_recession_data[['GDP', 'INDPRO', 'CPI', 'Housing_Index']].mean())
```

It helps compare how these indicators behave in stable economic conditions.


*   If the average GDP is lower during recessions, it suggests economic contraction.
*   Lower INDPRO values in recessions show that factories produce less due to weaker demand.

*   CPI fluctuations indicate how prices react to economic downturns.
*   A lower housing index means housing markets slow down during downturns.

# GDP & Industrial Production Over Time

```python
plt.figure(figsize=(12,6))
plt.plot(df1.index, df1['GDP'], label="GDP Growth (%)", marker='o', linestyle='-')
plt.plot(df1.index, df1['INDPRO'], label="Industrial Production", marker='s', linestyle='--', color='red')
plt.xlabel("Time (Index)")
plt.ylabel("Value")
plt.title("GDP Growth & Industrial Production Over Time")
plt.legend()
plt.grid()
plt.show()
```

# Correlation Between Economic Indicators

```python
plt.figure(figsize=(10,6))
sns.heatmap(df1.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.9)
plt.title("Correlation Between Economic Indicators")
plt.show()
```

Computes the correlation matrix of all numerical columns in df1.

The correlation values range from -1 to 1:

*   1.0 â†’ Perfect positive correlation (two variables increase together).
*   -1.0 â†’ Perfect negative correlation (one variable increases, the other decreases).
*   0.0 â†’ No correlation.


# Interest Rates During Recession vs. Non-Recession

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df1['Recession'], y=df1['Rate'], palette="coolwarm")
plt.xlabel("Recession (0 = No, 1 = Yes)")
plt.ylabel("Interest Rate (%)")
plt.title("Interest Rates During Recession vs. Non-Recession")
plt.show()
```

If interest rates drop during recessions, it indicates government intervention to stabilize the economy.

# Visualizing Recession Periods Over GDP

```python
plt.figure(figsize=(12,6))
plt.plot(df1.index, df1['GDP'], label="GDP Growth", color="blue")
plt.fill_between(df1.index, df1['GDP'], where=(df1['Recession'] == 1), color="red", alpha=0.3, label="Recession Period")
plt.xlabel("Time (Index)")
plt.ylabel("GDP Growth (%)")
plt.title("GDP Growth with Recession Periods Highlighted")
plt.legend()
plt.show()
```

Red-shaded areas mark recession periods, making it easy to see GDP drops.

# **Summery**

A recession is an economic situation that arrives when the circulation of money in the economy is low for two consecutive quarters. Recession is calculated and analyzed according to the growth in GDP, the growth in the unemployment rate, and the growth in consumer spending rate.
